```python
# ðŸŽµ Music Analytics Dashboard (using Pandas + NumPy)
# Project by shaurya Chawla


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create a dataset (Artists, Albums, Genres, Sales, Ratings)
# In real projects, this could come from a CSV or API
data = {
    'Artist': ['Tool','Pink Floyd','Metallica','Daft Punk','Michael Jackson','Tool','Beatles','Adele'],
    'Album': ['Lateralus','The Wall','Black Album','Discovery','Thriller','Fear Inoculum','Abbey Road','21'],
    'Genre': ['Prog Rock','Rock','Metal','Electronic','Pop','Prog Rock','Rock','Pop'],
    'Sales': [3.0, 30.0, 16.0, 2.5, 66.0, 0.5, 31.0, 22.0],   # sales in millions
    'Rating': [9.5, 9.7, 8.9, 8.5, 9.8, 8.7, 9.6, 9.0]
}

# Step 3: Convert to Pandas DataFrame
df = pd.DataFrame(data)
print("ðŸŽ¶ Full Dataset:\n")
print(df)


# ðŸŽ¯ TASK 1: Sort by Genre
print("\nðŸ“Œ Albums Sorted by Genre:\n")
print(df.sort_values(by='Genre'))


# ðŸŽ¯ TASK 2: Top 5 Albums by Sales
print("\nðŸ”¥ Top 5 Albums by Sales:\n")
print(df.sort_values(by='Sales', ascending=False).head(5))


# ðŸŽ¯ TASK 3: Mean Sales
mean_sales = df['Sales'].mean()
print("\nðŸ“Š Mean Album Sales: ", mean_sales, "million")


# ðŸŽ¯ TASK 4: Group by Genre (average sales per genre)
print("\nðŸŽ¼ Average Sales by Genre:\n")
print(df.groupby('Genre')['Sales'].mean())


# ðŸŽ¯ TASK 5: Save DataFrame to CSV
df.to_csv("music_dashboard.csv", index=False)
print("\nâœ… Data saved to music_dashboard.csv")


# ðŸŽ¯ TASK 6: Visualization - Bar Plot
# Group by Genre and sum sales
genre_sales = df.groupby('Genre')['Sales'].sum()

# Plot
genre_sales.plot(kind='bar', color='skyblue', title="ðŸŽµ Total Sales by Genre")
plt.ylabel("Sales (millions)")
plt.xlabel("Genre")
plt.show()
```
